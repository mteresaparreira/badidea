```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
superBAD_df = pd.read_excel('../../data/model_stat_features_ohe_norm.xlsx')
superBAD_df
```

```python
superBAD_df = pd.read_csv('../../data/30_1_data.csv')
#print unique participant_id and count

print(len(superBAD_df['participant_id'].unique()))
superBAD_df
```

### Train Test split

```python
features = superBAD_df.iloc[:, 4:]
target_class = superBAD_df.iloc[:, 2].values
target_class = target_class.astype('int')
# target_class.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target_class, test_size = 0.20, random_state = 42)
```

### Random Forests

```python
classifier = RandomForestClassifier(random_state = 42)
classifier.fit(X_train, y_train)

y_predict = classifier.predict(X_test)

print(classification_report(y_test, y_predict))
```

```python
classifier.get_params()
```

### GridSearch + RandomForests

```python
param_grid = {'n_estimators': [100, 200, 300, 400, 500],
    'criterion': ['gini','entropy', 'log_loss'],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True, False],
    'oob_score': [True, False]
    }

grid_search = GridSearchCV(RandomForestClassifier(random_state = 42), param_grid=param_grid, refit= True, verbose = True)
grid_search.fit(X_train, y_train)
```

```python
# print best parameter after tuning
print("Best Parameters: ", grid_search.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid_search.best_estimator_)
```

```python
grid_predictions = grid_search.predict(X_test)
  
# print classification report
print(classification_report(y_test, grid_predictions))
```
