```python
import numpy as np
import pandas as pd

from sklearn import svm
from sklearn.linear_model import LogisticRegression

from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report
```

```python
superBAD_df = pd.read_excel('../data/model_stat_features_ohe.xlsx')
superBAD_df
```

### Train Test split

```python
features = superBAD_df.iloc[:, 4:]
target_class = superBAD_df.iloc[:, 1].values
target_class = target_class.astype('int')
# target_class.shape
```

```python
scaler = StandardScaler()
features_norm = scaler.fit_transform(features)
features_norm = pd.DataFrame(features_norm, columns = features.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_norm, target_class, test_size = 0.20, random_state = 42)
```

### LogisticRegression

```python
log_classifier = LogisticRegression()

log_classifier.fit(X_train, y_train)

y_predict = log_classifier.predict(X_test)
print(classification_report(y_test, y_predict))
```
