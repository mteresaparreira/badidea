```python
#imports
import numpy as np
import pandas as pd
```

```python
#go to ../data/30_1_data and open
df = pd.read_csv('../data/30_1_data.csv')
print(df.head())
```

```python
#now, go participant by participant, response_video by response_video, and get the class for that video in new df
ground_truth = pd.DataFrame(columns=['participant_id', 'response_video', 'class'])
for participant in df['participant_id'].unique():
    df_part = df[df['participant_id'] == participant]
    for response_video in df_part['response_video'].unique():
        df_part_resp = df_part[df_part['response_video'] == response_video]
        ground_truth = pd.concat([ground_truth, pd.DataFrame({'participant_id': [participant], 'response_video': [response_video], 'class': [df_part_resp['class'].values[0]]})])

print(ground_truth.shape)
print(ground_truth.head())
ground_truth.to_csv('badidea_ground_truth.csv', index=False)
```
